```python
# CONFIG
MODEL_NAME = "meta-llama/Llama-3.1-8B"
REMOTE = True
```

```python
from nnterp import StandardizedTransformer

model = StandardizedTransformer(
    MODEL_NAME, 
    device_map="auto", 
    dispatch=not REMOTE,
    allow_dispatch=not REMOTE,
    check_renaming= not REMOTE,
    remote=False
)
```

## Chart

**Activation Patching Parameters:**

- `src_prompt` is your source prompt.
- `tgt_prompt` is the target prompt.
- `src_pos` is the position of the source token in the source prompt.
- `tgt_pos` is the position of the target token in the target prompt.
- `tgt_freeze` is a list of positions in the target prompt to freeze during patching.

```python
# PARAMETERS
src_prompt = """amount: cantidad, win: gañar, dreams:"""
tgt_prompt = """dog: chien, run: courir, flowers:"""
src_pos = [-1]
tgt_pos = [-1]
tgt_freeze = []
```

```python
from nnsightful import activation_patching

ap_data = activation_patching(
    model,
    src_prompt,
    tgt_prompt,
    src_pos,
    tgt_pos,
    tgt_freeze,
    remote=REMOTE,
)

ap_data.display()
```
