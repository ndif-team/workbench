```python
# CONFIG
MODEL_NAME = "meta-llama/Llama-3.1-8B"
REMOTE = True
```

```python
from nnsight import LanguageModel

model = LanguageModel(MODEL_NAME, device_map="auto", dispatch=not REMOTE)
```

## Chart

```python
# PARAMETERS
src_prompt = """amount: cantidad, win: gañar, dreams:"""
tgt_prompt = """dog: chien, run: courir, flowers:"""
src_pos = [-1]
tgt_pos = [-1]
tgt_freeze = []
```

```python
from nnsightful import activation_patching

ap_data = activation_patching.activation_patching(
    model,
    src_prompt,
    tgt_prompt,
    src_pos,
    tgt_pos,
    tgt_freeze,
    remote=REMOTE
)

ap_data.display()
```
