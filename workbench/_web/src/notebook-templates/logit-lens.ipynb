```python
# CONFIG
MODEL_NAME = "meta-llama/Llama-3.1-8B"
REMOTE = True
```

```python
from nnsight import LanguageModel

model = LanguageModel(MODEL_NAME, device_map="auto", dispatch=not REMOTE)
```

## Chart

**Logit Lens Parameters:**

- `prompt` is the input text to analyze.
- `topk` is the number of top predictions to show per layer.
- `include_entropy` controls whether entropy data is included in the visualization.

```python
# PARAMETERS
prompt = "The capital of France is"
topk = 5
include_entropy = True
```

```python
from nnsightful import logit_lens

# VISUALIZATION
ll_data = logit_lens(
    model,
    prompt,
    top_k=topk,
    include_entropy=include_entropy,
    remote=REMOTE
)

ll_data.display()
```
