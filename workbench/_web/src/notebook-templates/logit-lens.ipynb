```python
# CONFIG
MODEL_NAME = "meta-llama/Llama-3.1-8B"
REMOTE = True
```

```python
from nnterp import StandardizedTransformer

model = StandardizedTransformer(
    MODEL_NAME, 
    device_map="auto", 
    dispatch=not REMOTE,
    allow_dispatch=not REMOTE,
    check_renaming= not REMOTE,
    remote=False
)
```

## Chart

**Logit Lens Parameters:**

- `prompt` is the text whose intermediate-layer predictions you want to decode.
- `top_k` is the number of top predictions kept per cell.
- `include_entropy` toggles per-cell entropy computation.

```python
# PARAMETERS
prompt = """The Eiffel Tower is located in the city of"""
top_k = 5
include_entropy = True
```

```python
from nnsightful import logit_lens

ll_data = logit_lens(
    model,
    prompt,
    top_k=top_k,
    include_entropy=include_entropy,
    remote=REMOTE,
)

ll_data.display()
```
